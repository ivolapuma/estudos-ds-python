<b>Padronização</b>

<a href="https://en.wikipedia.org/wiki/Standard_score">Standardize</a> generally means changing the values so that the distribution standard deviation from the mean equals one. It outputs something very close to a normal distribution. Scaling is often implied.

```python
# média de uma série de números
def media(numeros):
    return sum(numeros) / len(numeros)

# variância de uma série de números
def variancia(numeros):
    m = media(numeros)
    vs = [(n - m)**2 for n in numeros]
    #return sum(vs) / (len(vs)-1) # formula variancia para amostra
    return sum(vs) / (len(vs)) # formula variancia StandardScaler()

# desvio padrão de uma série de números
def desvio_padrao(numeros):
    v = variancia(numeros)
    return v**(1/2)

# fórmula de padronização
def padroniza_valor(valor, media, desvio_padrao):
    return (valor - media) / desvio_padrao

# padronização de um numéro de uma série
def padronizar(numero, numeros):
    return padroniza_valor(numero, media(numeros), desvio_padrao(numeros))
```
