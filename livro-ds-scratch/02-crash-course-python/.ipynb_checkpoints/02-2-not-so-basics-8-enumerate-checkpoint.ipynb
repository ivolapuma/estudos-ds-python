```python
# iterando sobre uma lista e usando seus elementos e indices...

document = [
    "nada", "menos", "de", "duas", "almas", "cada", "criatura", "humana", "traz", "duas", 
    "almas", "consigo", "uma", "que", "olha", "de", "dentro", "para", "fora", "outra", 
    "que", "olha", "de", "fora", "para", "entro"   
]

# modo nao Pythonico
print("modo 1 - nao Pythonico")
for i in range(len(document)):
    word = document[i]
    print(i, word)
print()
    
# tambem nao Pythonico
print("modo 2 - nao Pythonico")
i = 0
for word in document:
    print(i, word)
    i += 1
print()
    
# uma solucao Pythonica eh o 'enumerate', que retorna tuplas (indice, elemento)
print("modo 3 - mais Pythonico")
for i, word in enumerate(document):
    print(i, word)
print()
```

```python
# para obter somente os indices

# modo nao Pythonico
for i in range(len(document)):
    print(i)

# modo mais Pythonico
for i, _ in enumerate(document):
    print(i)
```
