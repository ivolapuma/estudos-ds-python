# Derivadas

```python
import math
```

```python
'''
Cálculo aproximado da derivada:

derivada de f(x) = f'(x) = limite (h->0): (f(x+h) - f(x)) / h

ex: f(x) = x**2
    f'(x) = 2*x
'''
def derivada(f, x, limite = 5):
    # f: funcao a ser derivada
    # x: ponto em que ocorre a derivada
    # limite: (opcional) número inteiro negativo potencia de 10 usado no cálculo aproximado da derivada
    if limite > 0: limite = limite * (-1)
    h = 10**limite
    return (f(x + h) - f(x)) / h
```

```python
# testando cálculo aproximado de derivada...

# derivada de f(x) = x^2 --> f'(x) = 2x | x = 2
print("x = 2 | f(x) = x^2 | f'(x) = 2x =", derivada(lambda x: x**2, 2))
# derivada de f(x) = x^2 --> f'(x) = 2x | x = 3
print("x = 3 | f(x) = x^2 | f'(x) = 2x =", derivada(lambda x: x**2, 3))

# derivada de f(x) = sin(x) --> f'(x) = cos(x) | x = 90º 
print("x = 90º | f(x) = sin(x) | f'(x) = cos(x) =", derivada(lambda x: math.sin(x), math.radians(90)))
# derivada de f(x) = sin(x) --> f'(x) = cos(x) | x = 45º 
print("x = 45º | f(x) = sin(x) | f'(x) = cos(x) =", derivada(lambda x: math.sin(x), math.radians(45)))

# derivada de f(x) = ln(x) --> f'(x) = 1/x, x>0 | x = 2
print("x = 2 | f(x) = ln(x) | f'(x) = 1/x (x>0) =", derivada(lambda x: math.log(x), 2))
# derivada de f(x) = ln(x) --> f'(x) = 1/x, x>0 | x = 4
print("x = 4 | f(x) = ln(x) | f'(x) = 1/x (x>0) =", derivada(lambda x: math.log(x), 4))
```

```python
import matplotlib.pyplot as plt
```

```python
# plotando f(x) = x**2 e f'(x)
f = lambda x: x**2

x = [i for i in range(0,10)]
y = [f(i) for i in x]
y1 = [derivada(f, i) for i in x]

plt.plot(x, y, 'b', y1, 'r')
plt.show()
```
